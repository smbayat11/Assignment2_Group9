```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

Data_weater = pd.read_csv("weather_data.csv")
Data_price = pd.read_csv("price_demand_data.csv")
Data_weater
```

### Changing Both Data_Frame Date column as same format.

```python
# Changing the Data_weater Date format, assign to original 

Data_weater["Date"] = pd.to_datetime(Data_weater["Date"], format = '%d/%m/%Y')
Data_weater["Date"]
```

```python
Data_weater
```

```python
# Data cleanning - split the SETTLEMENTDATE and coverted Date format
Data_price[['Date','Time']]=Data_price['SETTLEMENTDATE'].str.split(' ', expand=True)
Data_price['Date'] = pd.to_datetime(Data_price["Date"])
Data_price["Date"].dt.date
```

### Conversion of totaldemand/0.5hrs into energy used per hrs and sorting maximum energy per day

```python
Data_price_Demand = Data_price.groupby(['Date'])['TOTALDEMAND'].max().reset_index()
Data_price_Demand
```

```python
Data_price_Demand['Max_Energy_used/day(kWh)'] = Data_price_Demand['TOTALDEMAND']*0.5
Data_price_Demand
```

### Merge both dataset into one Data_Frame & index by date

```python
Df_Demand_Weater = Data_weater.merge(Data_price_Demand, how = 'inner', on = 'Date')
#Df_Demand_Weater = Df_Demand_Weater.set_index('Date')
Df_Demand_Weater
```

```python
# Average daily temperature

temp = Df_Demand_Weater.loc[:, "Minimum temperature (°C)":"Maximum temperature (°C)"]
Df_Demand_Weater['Average_temp_/day(°C)'] = temp.mean(axis = 1)
Df_Demand_Weater
```

```python
Df_Demand_Weater['Minimum temperature (°C)'].mean()
```

```python
Df_Demand_Weater['Maximum temperature (°C)'].mean()
```

```python
# Graph Average_temp/day Vs daily Max energy used, pearson correlation

Df_Demand_Weater.plot.scatter (x = 'Average_temp_/day(°C)', y = 'Max_Energy_used/day(kWh)', alpha = 0.8)
plt.show()

print("Pearson r is ",Df_Demand_Weater['Average_temp_/day(°C)'].corr(Df_Demand_Weater['Max_Energy_used/day(kWh)']))
```

```python
Df_Demand_Weater['Max_Energy_used/day(kWh)'].describe()
```

## Average temp in a specific_session Vs maximum daily energy used, & pearson correlation

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

ax1 = axes[0][0]
summer_start = '2020-12-01'
summer_end =  '2021-02-28'
Df_summer = Df_Demand_Weater.loc[(Df_Demand_Weater['Date'] >= summer_start) & (Df_Demand_Weater['Date'] <= summer_end)]
summer.plot.scatter (x = 'Average_temp_/day(°C)', y = 'Max_Energy_used/day(kWh)', title="summer", ax = ax1, lw = 1, c = 'r')
print("Pearson r is ",Df_summer['Average_temp_/day(°C)'].corr(Df_summer['Max_Energy_used/day(kWh)']), "in summer time")


ax2 = axes[0][1]
autumn_start = '2021-3-01'
autumn_end = '2021-5-31'
Df_autumn = Df_Demand_Weater.loc[(Df_Demand_Weater['Date'] >= autumn_start) & (Df_Demand_Weater['Date'] <= autumn_end)]
autumn.plot.scatter (x = 'Average_temp_/day(°C)', y = 'Max_Energy_used/day(kWh)',title="autumn", ax = ax2, lw = 1, c = 'b')
print("Pearson r is ",Df_autumn['Average_temp_/day(°C)'].corr(Df_autumn['Max_Energy_used/day(kWh)']), "in Autumn time")


ax3 = axes[1][0]
winter_start = '2021-06-01'
winter_end = '2021-08-31'
Df_winter = Df_Demand_Weater.loc[(Df_Demand_Weater['Date'] >= '2021-06-01') & (Df_Demand_Weater['Date'] <= '2021-08-31') ]
winter.plot.scatter (x = 'Average_temp_/day(°C)', y = 'Max_Energy_used/day(kWh)',title="winter", ax = ax3, lw = 1, c = 'k')
print("Pearson r is ",Df_winter['Average_temp_/day(°C)'].corr(Df_winter['Max_Energy_used/day(kWh)']), "in winter time")


plt.subplots_adjust(left=0.1,bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
plt.show()
```
